```python
# @title Testing code directly from the Tinytroupe Repo
from tinytroupe.examples import create_lisa_the_data_scientist

lisa = create_lisa_the_data_scientist() # instantiate a Lisa from the example builder
lisa.listen_and_act(input("Ask Lisa a Question: "))
```

```python
# @title Simple world creation and chat example

from tinytroupe.factory import TinyPersonFactory
from tinytroupe.environment import TinyWorld

factory_0 = TinyPersonFactory(context="A hospital in New York")
jimmy_0 = factory_0.generate_person("Create a New York person that is a data science graduate student, likes pets and nature and loves heavy metal.")

world = TinyWorld("Chat Room", [lisa, jimmy_0])
world.make_everyone_accessible()
lisa.listen("Talk to jimmy to know teach him what exactly a data scientist does in their career day to day life")
world.run(4)
```

```python
# @title Further Exploring the Tinytroupe Package with custom agents and situation

from tinytroupe.agent import TinyPerson

factory = TinyPersonFactory(context="A career center in NYC for Data Scientist and Software Engineers")
jimmy = factory.generate_person("Create Jimmy Neutron, who is a person that is an experienced data scientist in NYC")
sharon = factory.generate_person("Create Sharon Glimps, who an experience recruiter for Google NYC and has been hiring Data Scientist for 5+ years")

bodrul = TinyPerson("Bodrul Jalal")

bodrul.define("age", 23)
bodrul.define("nationality", "United States")
bodrul.define("occupation", {
              "title": "Data Scientist Master's Student",
              "organization": "Pace University",
              "description":
              """
              You are a data scientist student pursuing a master's degree in data science at Pace University. You are set to
              graduate in one year and are looking to get a job but don't know what field or position to apply for.
              """})

bodrul.define("behaviors", {"routines": ["Every morning, you wake up, do some yoga, and check your emails."]})

bodrul.define("personality",
                      {"traits": [
                          "You are curious and love to learn new things.",
                          "You are analytical and like to solve problems.",
                          "You are friendly and enjoy working with others.",
                          "You don't give up easily, and always try to find a solution. However, sometimes you can get frustrated when things don't work as expected."
                    ]})

bodrul.define("preferences",
                      {"interests": [
                        "Artificial intelligence and machine learning.",
                        "Natural language processing and conversational agents.",
                        "Search engine optimization and user experience.",
                        "Watching tv shows, especially detective and crime solving shows.",
                        "Learning random facts and how things work"
                      ]})

bodrul.define("skills",
                      [
                        "You are able to use various data analysis and machine learning tools, such as pandas, scikit-learn, TensorFlow, and Azure ML.",
                        "You are familiar with SQL and Python, but struggle with R."
                      ])

world_2 = TinyWorld("Coffe Chat", [jimmy, sharon, bodrul])
world_2.make_everyone_accessible()
bodrul.listen("Talk to jimmy and sharon to learn how to land a data science job and what steps you should take to better your chances of success.")
world_2.run(4)
```

```python
# @title Testing the summary extraction function

from tinytroupe.extraction import ResultsExtractor

summary = ResultsExtractor().extract_results_from_world(
    world_2,
    extraction_objective="Summarize key points and ideas from each person in the world",
    verbose=False
)

print ("="*30)

for key, val in summary.items():
  print("{}'s summary: {}".format(key, val))
  print("=" * 30)
```
